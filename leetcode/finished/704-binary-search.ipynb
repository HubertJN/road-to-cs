## Binary Search

Given an array of integers `nums` which is sorted in ascending order, and an integer `target`, write a function to search for `target` in `nums`.

If `target` exists, return its index. Otherwise, return `-1`.

You must write an algorithm with **O(log n)** runtime complexity.

### Example 1

**Input:**
`nums = [-1,0,3,5,9,12], target = 9`

**Output:**
`4`

**Explanation:**
`9` exists in `nums` and its index is `4`.

### Example 2

**Input:**
`nums = [-1,0,3,5,9,12], target = 2`

**Output:**
`-1`

**Explanation:**
`2` does not exist in `nums`, so return `-1`.

### Constraints

* `1 <= nums.length <= 10^4`
* `-10^4 < nums[i], target < 10^4`
* All the integers in `nums` are unique.
* `nums` is sorted in ascending order.

```python
def brute_search(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1

def binary_search_iterative(nums, target):
    def bsr(nums, target, left, right):
        if left > right:
            return -1
        mid = (left + right) // 2
        if nums[mid] < target:
            return bsr(nums, target, mid + 1, right)
        elif nums[mid] > target:
            return bsr(nums, target, left, mid - 1)
        else:
            return mid

    return bsr(nums, target, 0, len(nums)-1)

def binary_search_iterative(nums, target):
    left = 0
    right = len(nums)-1
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] < target:
            left = mid + 1
        elif nums[mid] > target:
            right = mid - 1
        else:
            return mid
    return -1
```

```python
import time
import statistics

def run_scaling_test(fn, sizes=None, batch_size=1000, samples=7):
    if sizes is None:
        sizes = [10, 100, 1_000, 10_000]

    print(f"{'n':>10} | {'median (us)':>15} | {'vs previous':>12}")
    print("-" * 45)

    previous_time = None

    for n in sizes:
        nums = list(range(n))
        target = nums[-1]

        sample_times = []

        for _ in range(samples):
            start = time.perf_counter()

            for _ in range(batch_size):
                fn(nums, target)

            total_time = time.perf_counter() - start
            time_per_call_us = (total_time / batch_size) * 1_000_000
            sample_times.append(time_per_call_us)

        median_time = statistics.median(sample_times)

        if previous_time is None:
            comparison = "N/A"
        else:
            comparison = f"{median_time / previous_time:.2f}x"

        print(f"{n:>10} | {median_time:>15.3f} | {comparison:>12}")
        previous_time = median_time
```

## Takeaway

The recursive and iterative binary searches are both **O(log n)**, but the recursive version is slower in Python because each recursive step creates a new function call and stack frame. That adds overhead for argument passing, returning values, and managing the call stack. The iterative version uses the same search logic while avoiding that overhead, so it is usually faster and also uses less space.

By contrast, **brute force is worse because it is O(n)**. In the worst case, it may need to check every element one by one before finding the target, or determine that it is not present. Binary search takes advantage of the fact that the array is sorted, so it can discard half of the remaining search space at each step.
