# 587. Erect the Fence

You are given an array `trees` where `trees[i] = [xi, yi]` represents the location of a tree in the garden.

You are asked to fence the entire garden using the minimum length of rope. The garden is well-fenced only if all the trees are enclosed.

Return the coordinates of trees that are exactly located on the fence perimeter. You may return the answer in any order.

## Example 1

**Input:** `trees = [[1,1],[2,2],[2,0],[2,4],[3,3],[4,2]]`  
**Output:** `[[1,1],[2,0],[4,2],[3,3],[2,4]]`

## Example 2

**Input:** `trees = [[1,2],[2,2],[4,2]]`  
**Output:** `[[4,2],[2,2],[1,2]]`

## Constraints

- `1 <= trees.length <= 3000`
- `trees[i].length == 2`
- `0 <= xi, yi <= 100`
- All the given positions are unique.


```python
from typing import List


class Solution:
    def outerTrees(self, trees: List[List[int]]) -> List[List[int]]:
        def cross(a, b, c):
            return (b[0] - a[0])*(c[1] - a[1]) - (b[1] - a[1])*(c[0] - a[0])

        points = sorted(map(tuple, trees))
        print(points)

        lower = []
        upper = []
        for tree in points:
            while len(lower) > 1 and cross(lower[-2], lower[-1], tree) < 0:
                lower.pop()
            while len(upper) > 1 and cross(upper[-2], upper[-1], tree) > 0:
                upper.pop()
            lower.append(tree)
            upper.append(tree)

        return [list(p) for p in set(upper + lower)]
```

```python
sol = Solution()

test_cases = [
    [[1, 1], [2, 2], [2, 0], [2, 4], [3, 3], [4, 2]],
    [[1, 2], [2, 2], [4, 2]],
]

for trees in test_cases:
    result = sol.outerTrees(trees)
    print(f"input={trees}")
    print(f"output={result}")
    print()
```
