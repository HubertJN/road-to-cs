# 232. Implement Queue using Stacks

Implement a first in first out (FIFO) queue using only two stacks. The implemented queue should support all the functions of a normal queue (`push`, `peek`, `pop`, and `empty`).

Implement the `MyQueue` class:

- `void push(int x)` Pushes element `x` to the back of the queue.
- `int pop()` Removes the element from the front of the queue and returns it.
- `int peek()` Returns the element at the front of the queue.
- `boolean empty()` Returns `true` if the queue is empty, `false` otherwise.

## Example 1

**Input:** `["MyQueue", "push", "push", "peek", "pop", "empty"]`, `[[], [1], [2], [], [], []]`  
**Output:** `[null, null, null, 1, 1, false]`

## Constraints

- `1 <= x <= 9`
- At most `100` calls will be made to `push`, `pop`, `peek`, and `empty`.
- All the calls to `pop` and `peek` are valid.


```python
class MyQueue:
    def __init__(self):
        self.stack = [[], []]
        self.input = 0
        self.output = 1

    def push(self, x: int) -> None:
        self.stack[self.input].append(x)

    def pop(self) -> int:
        top = self.peek()
        self.stack[self.output].pop()
        return top

    def peek(self) -> int:
        if not self.stack[self.output]:
            for _ in range(len(self.stack[self.input])):
                self.stack[self.output].append(self.stack[self.input].pop())
        return self.stack[self.output][-1]

    def empty(self) -> bool:
        return not self.stack[0] and not self.stack[1]

```

```python
queue = MyQueue()

queue.push(1)
queue.push(2)
print(queue.peek())
print(queue.pop())
print(queue.empty())
```
